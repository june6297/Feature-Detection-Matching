```python
import os
import random
import shutil
from sklearn.model_selection import train_test_split

def split_dataset(src_dir_normal, src_dir_damaged, dst_dir, test_size=0.2):
    # 디렉토리 생성
    os.makedirs(os.path.join(dst_dir, 'train', 'normal'), exist_ok=True)
    os.makedirs(os.path.join(dst_dir, 'train', 'damaged'), exist_ok=True)
    os.makedirs(os.path.join(dst_dir, 'test', 'normal'), exist_ok=True)
    os.makedirs(os.path.join(dst_dir, 'test', 'damaged'), exist_ok=True)

    # 파일 목록 가져오기
    normal_files = os.listdir(src_dir_normal)
    damaged_files = os.listdir(src_dir_damaged)

    # 훈련/테스트 세트 분할
    normal_train, normal_test = train_test_split(normal_files, test_size=test_size, random_state=42)
    damaged_train, damaged_test = train_test_split(damaged_files, test_size=test_size, random_state=42)

    # 파일 복사
    for file in normal_train:
        shutil.copy(os.path.join(src_dir_normal, file), os.path.join(dst_dir, 'train', 'normal', file))
    for file in normal_test:
        shutil.copy(os.path.join(src_dir_normal, file), os.path.join(dst_dir, 'test', 'normal', file))
    for file in damaged_train:
        shutil.copy(os.path.join(src_dir_damaged, file), os.path.join(dst_dir, 'train', 'damaged', file))
    for file in damaged_test:
        shutil.copy(os.path.join(src_dir_damaged, file), os.path.join(dst_dir, 'test', 'damaged', file))

# 사용 예
src_dir_normal = "/content/drive/MyDrive/KeyPoint/images2"
src_dir_damaged = "/content/drive/MyDrive/KeyPoint/images4"
dst_dir = "/content/drive/MyDrive/KeyPoint/dataset_split"
split_dataset(src_dir_normal, src_dir_damaged, dst_dir)
```
